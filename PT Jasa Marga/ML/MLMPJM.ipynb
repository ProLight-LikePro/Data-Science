```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# load the data
df = pd.read_csv("DPJM.csv")
```

```python
# create scatter plot
%matplotlib notebook
df.plot(kind = 'scatter', x = 'DAR', y = 'NPM')
plt.show()
```

```python
# Test train split for supervised training & visualization
X_train, X_test, y_train, y_test = train_test_split(df.DAR, df.NPM)

plt.scatter(X_train, y_train, label = 'Training Data', color = 'r')
plt.scatter(X_test, y_test, label = 'Testing Data', color = 'g')
plt.legend()
plt.title("Test Train Split")
plt.show()
```

```python
# Create linear model and train it
LR = LinearRegression()
LR.fit(X_train.values.reshape(-1,1), y_train.values)
```

```python
# Use model to predict on Test data
prediction = LR.predict(X_test.values.reshape(-1,1))

# plot prediction line against actual test data
plt.plot(X_test, prediction, label = 'Linear Regression', color = 'b')
plt.scatter(X_test, y_test, label = 'Actual Test Data', color = 'g')
plt.legend()
plt.show()
```

```python
# predict specific value
LR.predict(np.array([[50], [45], [0]]))
```

```python
# Score the model
LR.score(X_test.values.reshape(-1, 1), y_test.values)
```
